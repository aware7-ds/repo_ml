# Space & Time Complexity

```python
x= [1,2,3,4,5]
print(5 in x)
```

```python
##  if len of list or num of elements in list ie.. if n increases,comparisons/computations also increases
## n is proportional to comparisons.. 
## time complexities -- o(n),o(n^2),o(log n),o(1)....

## log(n) <= n
## better soln : log n comparisons than n comparisons
## Binary Search takes O(Log n) comparisons -- sorted list;recursion; (l+r)/2
```

```python
x = [21, 47, 32, 12, 98, 3, 78, 45, 7, 8, 43 ]    ## Binary Sort Algorithm
v = 45
x.sort()
print(x)
import math
def Binary_search(arr,l,r,v):
    if r>=l:
        mid = math.floor((l+r)/2)
        if v == arr[mid]:
            return ('value found', arr[mid])
        elif v > arr[mid]:
            return Binary_search(arr,mid+1,r,v)
        else:
            return Binary_search(arr,l,mid-1,v)
    else:
        return 'Value not found'
        
    
Binary_search(x,0,len(x)-1,v)    
```

```python
## Finding common elements in two lists ---  O(n*m)

import random
l1 = list(range(20))   ## n elements
random.shuffle(l1)    ## shuffling 20 values using random.shuffle()
l2 = list(range(7))    ## m elements
random.shuffle(l2)
count = 0
for i in l1:              ## time complexity : O(n*m); space complexity = O(1) -Ram; time reduction is imp
    for j in l2:
        if i == j:
            count += 1
print(' count: ',count)
```

```python
## Finding common elements in two lists ---  using Dict  --- O(m+n) ; O(m+n <= O(m*n))

## Dictionaries - similar - hash table/ hash map
## Dictionaries traverse very faster ie.. directly to the key unlike linear/binary searches; it uses Hash Func
## time complexity - O(1);

import random
l1 = list(range(20))   ## n elements
random.shuffle(l1)    ## shuffling 20 values using random.shuffle()
l2 = list(range(7))    ## m elements
random.shuffle(l2)

dict = {}
x = min(l1,l2)   ## Store the smallest list(lengthwise) in a dict so that each element acts as a key
print(x)

for i in x:
    dict[i]= 1  ## Value is not imp; key is imp; space complexity = O(m); time complexity = O(m) here; l2 - m elements; loop
print(dict)
count = 0
for i in l1:
    if dict.get(i):
        count += 1   ## time complexity = O(n); l1- n elemnts; loop

print(count)     ## as a whole , time complexity = O(m+n)

 
```

```python
## CDF - cumulative distribution Func
## CDF -> 0-1; we can get percentage of accuracy as a whole
```

```python
dataset_s = dataset[dataset['species'] == 'setosa']
counts,bin_edges = np.histogram(dataset_s['petal_length'],bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print('**********************')
print(bin_edges)

## compute CDF

cdf = np.cumsum(pdf)    ## cumsum()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

dataset_ver = dataset[dataset['species'] == 'versicolor']
counts,bin_edges = np.histogram(dataset_ver['petal_length'],bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print('**********************')
print(bin_edges)

## compute CDF

cdf = np.cumsum(pdf)    ## cumsum()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

dataset_vir = dataset[dataset['species'] == 'virginica']
counts,bin_edges = np.histogram(dataset_vir['petal_length'],bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print('**********************')
print(bin_edges)

## compute CDF

cdf = np.cumsum(pdf)    ## cumsum()
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf,)

plt.grid()
plt.show()
```

```python
##  CDF curves ranges from 0-1
## intake from CDF curve/ what to understand from CDF curve: if we take setosa(blue-orange) , 80% of setosa flowers have 
## petal length(x-axis) less than 1.6
## intake from PDF curve : 20% of setosa flowers have petal length between 1.4 to 1.6
```

```python
## we can get some info from mean n std(variance) without drawin a plot

print('Mean:')
print(np.mean(dataset_s['petal_length']))
print(np.mean(dataset_ver['petal_length']))
print(np.mean(dataset_vir['petal_length']))

print('\nStd:')
print(np.std(dataset_s['petal_length']))
print(np.std(dataset_ver['petal_length']))
print(np.std(dataset_vir['petal_length']))

## outlier: if atleast one datapoint/observations is error, whole mean n std vl vary drastically

print('\nOutlier:')
print(np.mean(np.append((dataset_s['petal_length']),50)))
```

```python
## here we can say that average petal lengths of setosa are less than other two - mean
##  setosa petals are less spread than other two(densely grouped)
## variance/spread/std - how wide petals are
## std = square root of variance
## if v have variance assume 0.5, then std of setosa will be mean +/- variance ie.. (0.9,1.9)- width of petals 
## Range: (1.46-0.17, 1.46, 1.46+0.17)  ## setosa

## Mean n Std vl vary if there is any outlier.. alternative is median for mean; median absolute deviation for std
```
