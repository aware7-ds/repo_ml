```python
l = [1,2,3,4,5]
def add_rec(l,size):
  if size == 0:
    p = 1
  else:
    p = l[size-1]*add_rec(l,size-1)
  return p
print(add_rec(l,len(l)))
```

```python
def get_sum(lst):
    sum = 0
    for i in lst:
        sum += i
    return(sum)
x = get_sum([1,2,3,4])
print('sum:',x)
        
```

```python
def hcf(a,b):
    '''computes hcf'''
    smallest = a if a<b else b
    h=1
    for i in range(1,smallest+1):
        if (a%i == 0) & (b%i ==0):
            h = i
    return h
y = hcf(7,8)
print('hcf:',y)
            
```
